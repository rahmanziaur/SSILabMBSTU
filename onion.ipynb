```python

import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 1
x = np.linspace(0, 1, 1000)

# Define the functions
functions = {
    '$y=x$': x,
    '$y=x^{2}$': x**2,
    '$y=x^{3}$': x**3,
    '$y=x^{4}$': x**4,
    '$y=x^{5}$': x**5,
    '$y=x^{6}$': x**6,
    '$y=x^{8}$': x**8,
    '$y=x^{10}$': x**10,
    '$y=\sqrt{x}$': np.sqrt(x),
    '$y=\sqrt[3]{x}$': x**(1/3),
    '$y=\sqrt[4]{x}$': x**(1/4),
    '$y=\sqrt[5]{x}$': x**(1/5),
    '$y=\sqrt[7]{x}$': x**(1/7),
    '$y=\sqrt[8]{x}$': x**(1/8),
    '$y=\sqrt[9]{x}$': x**(1/9),
    '$y=\sqrt[10]{x}$': x**(1/10)
}

# Plot the functions
plt.figure(figsize=(8, 8))
for label, y_val in functions.items():
    plt.plot(x, y_val, label=label)

# Add labels, grid, and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of Power and Root Functions (Onion Graph)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='lower right', fontsize='small')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box') # Ensure equal aspect ratio

# Show the plot
plt.show()
```

```python

%%writefile main.go
// paste Go code here (produces onion_go.png)

package main

import (
    "image/color"
    "math"
    "gonum.org/v1/plot"
    "gonum.org/v1/plot/plotter"
    "gonum.org/v1/plot/vg"
)

func main() {
    p := plot.New()
    p.Title.Text = "Power & Root Functions (Onion Graph)"
    p.X.Label.Text = "x"
    p.Y.Label.Text = "y"

    functions := []struct {
        power float64
        name  string
        color color.Color
    }{
        {1, "x", color.RGBA{31, 119, 180, 255}},
        {2, "x", color.RGBA{255, 127, 14, 255}},
        {3, "x", color.RGBA{44, 160, 44, 255}},
        {0.5, "x", color.RGBA{188, 189, 34, 255}},
        {1.0 / 3, "x", color.RGBA{23, 190, 207, 255}},
        // add more...
    }

    for _, fn := range functions {
        pts := make(plotter.XYs, 1000)
        for i := 0; i < 1000; i++ {
            x := float64(i) / 999
            pts[i].X = x
            pts[i].Y = math.Pow(x, fn.power)
        }

        line, _ := plotter.NewLine(pts)
        line.Color = fn.color
        line.Width = vg.Points(1.5)
        p.Add(line)
        p.Legend.Add(fn.name, line)
    }

    p.Save(6*vg.Inch, 6*vg.Inch, "onion_go.png")
}
```

```python

%%writefile onion_plot.go
package main

import (
    "fmt"
    "math"
    "os"
)

func main() {
    width, height := 600.0, 600.0
    margin := 40.0

    // Output SVG
    f, err := os.Create("onion_plot.svg")
    if err != nil {
        panic(err)
    }
    defer f.Close()

    // Write SVG header
    fmt.Fprintf(f, `<svg width="%v" height="%v" xmlns="http://www.w3.org/2000/svg">`, width, height)

    // Draw grid
    fmt.Fprintf(f, `<rect x="0" y="0" width="%v" height="%v" fill="white"/>`, width, height)
    fmt.Fprintf(f, `<line x1="%v" y1="%v" x2="%v" y2="%v" stroke="black"/>`, margin, height-margin, width-margin, height-margin)
    fmt.Fprintf(f, `<line x1="%v" y1="%v" x2="%v" y2="%v" stroke="black"/>`, margin, margin, margin, height-margin)

    // Functions to plot
    functions := []struct {
        name string
        fn   func(float64) float64
        color string
    }{
        {"y=x", func(x float64) float64 { return x }, "black"},
        {"y=x^2", func(x float64) float64 { return math.Pow(x, 2) }, "red"},
        {"y=x^3", func(x float64) float64 { return math.Pow(x, 3) }, "blue"},
        {"y=x^4", func(x float64) float64 { return math.Pow(x, 4) }, "green"},
        {"y=x^5", func(x float64) float64 { return math.Pow(x, 5) }, "purple"},
        {"y=x^6", func(x float64) float64 { return math.Pow(x, 6) }, "orange"},
        {"y=x^8", func(x float64) float64 { return math.Pow(x, 8) }, "brown"},
        {"y=x^10", func(x float64) float64 { return math.Pow(x, 10) }, "cyan"},

        {"y=sqrt(x)", func(x float64) float64 { return math.Sqrt(x) }, "darkblue"},
        {"y=x^(1/3)", func(x float64) float64 { return math.Pow(x, 1.0/3.0) }, "darkgreen"},
        {"y=x^(1/4)", func(x float64) float64 { return math.Pow(x, 1.0/4.0) }, "darkred"},
        {"y=x^(1/5)", func(x float64) float64 { return math.Pow(x, 1.0/5.0) }, "teal"},
        {"y=x^(1/7)", func(x float64) float64 { return math.Pow(x, 1.0/7.0) }, "magenta"},
        {"y=x^(1/8)", func(x float64) float64 { return math.Pow(x, 1.0/8.0) }, "gray"},
        {"y=x^(1/9)", func(x float64) float64 { return math.Pow(x, 1.0/9.0) }, "pink"},
        {"y=x^(1/10)", func(x float64) float64 { return math.Pow(x, 1.0/10.0) }, "gold"},
    }

    // Plot each function
    for _, fn := range functions {
        fmt.Fprintf(f, `<path fill="none" stroke="%s" stroke-width="2" d="`, fn.color)

        for i := 0; i <= 1000; i++ {
            x := float64(i) / 1000
            y := fn.fn(x)

            // Scale to SVG coordinates
            sx := margin + x*(width-2*margin)
            sy := height - (margin + y*(height-2*margin))

            if i == 0 {
                fmt.Fprintf(f, "M %f %f ", sx, sy)
            } else {
                fmt.Fprintf(f, "L %f %f ", sx, sy)
            }
        }
        fmt.Fprint(f, `"/>`)
    }

    fmt.Fprintln(f, "</svg>")
}
```

```python

%%writefile onion_plot_with_legend.go
package main

import (
    "fmt"
    "math"
    "os"
)

func main() {
    width, height := 800.0, 600.0
    margin := 50.0

    // Open SVG file
    f, err := os.Create("onion_plot_with_legend.svg")
    if err != nil {
        panic(err)
    }
    defer f.Close()

    fmt.Fprintf(f, `<svg width="%v" height="%v" xmlns="http://www.w3.org/2000/svg">`, width, height)

    // Draw background
    fmt.Fprintf(f, `<rect x="0" y="0" width="%v" height="%v" fill="white"/>`, width, height)

    // Functions to plot
    functions := []struct {
        name  string
        fn    func(float64) float64
        color string
    }{
        {"y=x", func(x float64) float64 { return x }, "black"},
        {"y=x^2", func(x float64) float64 { return math.Pow(x, 2) }, "red"},
        {"y=x^3", func(x float64) float64 { return math.Pow(x, 3) }, "blue"},
        {"y=x^4", func(x float64) float64 { return math.Pow(x, 4) }, "green"},
        {"y=x^5", func(x float64) float64 { return math.Pow(x, 5) }, "purple"},
        {"y=x^6", func(x float64) float64 { return math.Pow(x, 6) }, "orange"},
        {"y=x^8", func(x float64) float64 { return math.Pow(x, 8) }, "brown"},
        {"y=x^10", func(x float64) float64 { return math.Pow(x, 10) }, "cyan"},
        {"y=sqrt(x)", func(x float64) float64 { return math.Sqrt(x) }, "darkblue"},
        {"y=x^(1/3)", func(x float64) float64 { return math.Pow(x, 1.0/3.0) }, "darkgreen"},
        {"y=x^(1/4)", func(x float64) float64 { return math.Pow(x, 1.0/4.0) }, "darkred"},
        {"y=x^(1/5)", func(x float64) float64 { return math.Pow(x, 1.0/5.0) }, "teal"},
        {"y=x^(1/7)", func(x float64) float64 { return math.Pow(x, 1.0/7.0) }, "magenta"},
        {"y=x^(1/8)", func(x float64) float64 { return math.Pow(x, 1.0/8.0) }, "gray"},
        {"y=x^(1/9)", func(x float64) float64 { return math.Pow(x, 1.0/9.0) }, "pink"},
        {"y=x^(1/10)", func(x float64) float64 { return math.Pow(x, 1.0/10.0) }, "gold"},
    }

    // Plot curves
    for _, fn := range functions {
        fmt.Fprintf(f, `<path fill="none" stroke="%s" stroke-width="2" d="`, fn.color)
        for i := 0; i <= 1000; i++ {
            x := float64(i) / 1000
            y := fn.fn(x)
            sx := margin + x*(width-2*margin-200) // leave 200px for legend
            sy := height - (margin + y*(height-2*margin))
            if i == 0 {
                fmt.Fprintf(f, "M %f %f ", sx, sy)
            } else {
                fmt.Fprintf(f, "L %f %f ", sx, sy)
            }
        }
        fmt.Fprint(f, `"/>`)
    }

    // Add legend
    legendX := width - 180.0
    legendY := margin
    boxSize := 15.0
    spacing := 25.0

    for i, fn := range functions {
        yPos := legendY + float64(i)*spacing
        // Color box
        fmt.Fprintf(f, `<rect x="%f" y="%f" width="%f" height="%f" fill="%s"/>`, legendX, yPos, boxSize, boxSize, fn.color)
        // Text label
        fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="14">%s</text>`, legendX+20, yPos+boxSize-2, fn.name)
    }

    // Add title
    fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="18" font-weight="bold">Graphs of Power and Root Functions (Onion Graph)</text>`, margin, margin-20)

    fmt.Fprintln(f, "</svg>")
}
```

```python

%%writefile onion_plot_advanced.go
package main

import (
    "fmt"
    "math"
    "os"
)

func main() {
    width, height := 800.0, 600.0
    margin := 60.0

    // Open SVG file
    f, err := os.Create("onion_plot_advanced.svg")
    if err != nil {
        panic(err)
    }
    defer f.Close()

    fmt.Fprintf(f, `<svg width="%v" height="%v" xmlns="http://www.w3.org/2000/svg">`, width, height)
    fmt.Fprintf(f, `<rect x="0" y="0" width="%v" height="%v" fill="white"/>`, width, height)

    // Draw light grid
    numGrid := 10
    for i := 0; i <= numGrid; i++ {
        x := margin + float64(i)*(width-2*margin)/float64(numGrid)
        y := height - (margin + float64(i)*(height-2*margin)/float64(numGrid))
        fmt.Fprintf(f, `<line x1="%f" y1="%f" x2="%f" y2="%f" stroke="#ddd" stroke-width="1"/>`, x, margin, x, height-margin)
        fmt.Fprintf(f, `<line x1="%f" y1="%f" x2="%f" y2="%f" stroke="#ddd" stroke-width="1"/>`, margin, y, width-margin, y)
    }

    // Functions to plot
    functions := []struct {
        name  string
        fn    func(float64) float64
        color string
        labelX float64 // relative position for label (0 to 1)
    }{
        {"y=x", func(x float64) float64 { return x }, "black", 0.9},
        {"y=x^2", func(x float64) float64 { return math.Pow(x, 2) }, "red", 0.8},
        {"y=x^3", func(x float64) float64 { return math.Pow(x, 3) }, "blue", 0.7},
        {"y=x^4", func(x float64) float64 { return math.Pow(x, 4) }, "green", 0.6},
        {"y=x^5", func(x float64) float64 { return math.Pow(x, 5) }, "purple", 0.5},
        {"y=x^6", func(x float64) float64 { return math.Pow(x, 6) }, "orange", 0.4},
        {"y=x^8", func(x float64) float64 { return math.Pow(x, 8) }, "brown", 0.3},
        {"y=x^10", func(x float64) float64 { return math.Pow(x, 10) }, "cyan", 0.25},
        {"y=sqrt(x)", func(x float64) float64 { return math.Sqrt(x) }, "darkblue", 0.8},
        {"y=x^(1/3)", func(x float64) float64 { return math.Pow(x, 1.0/3.0) }, "darkgreen", 0.7},
        {"y=x^(1/4)", func(x float64) float64 { return math.Pow(x, 1.0/4.0) }, "darkred", 0.6},
        {"y=x^(1/5)", func(x float64) float64 { return math.Pow(x, 1.0/5.0) }, "teal", 0.5},
        {"y=x^(1/7)", func(x float64) float64 { return math.Pow(x, 1.0/7.0) }, "magenta", 0.4},
        {"y=x^(1/8)", func(x float64) float64 { return math.Pow(x, 1.0/8.0) }, "gray", 0.35},
        {"y=x^(1/9)", func(x float64) float64 { return math.Pow(x, 1.0/9.0) }, "pink", 0.3},
        {"y=x^(1/10)", func(x float64) float64 { return math.Pow(x, 1.0/10.0) }, "gold", 0.25},
    }

    // Plot curves
    for _, fn := range functions {
        fmt.Fprintf(f, `<path fill="none" stroke="%s" stroke-width="2" d="`, fn.color)
        for i := 0; i <= 1000; i++ {
            x := float64(i) / 1000
            y := fn.fn(x)
            sx := margin + x*(width-2*margin)
            sy := height - (margin + y*(height-2*margin))
            if i == 0 {
                fmt.Fprintf(f, "M %f %f ", sx, sy)
            } else {
                fmt.Fprintf(f, "L %f %f ", sx, sy)
            }
        }
        fmt.Fprint(f, `"/>`)

        // Label position along the curve
        lx := margin + fn.labelX*(width-2*margin)
        ly := height - (margin + fn.fn(fn.labelX)*(height-2*margin))
        fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="12" fill="%s">%s</text>`, lx+5, ly-2, fn.color, fn.name)
    }

    // Axes
    fmt.Fprintf(f, `<line x1="%f" y1="%f" x2="%f" y2="%f" stroke="black" stroke-width="2"/>`, margin, height-margin, width-margin, height-margin)
    fmt.Fprintf(f, `<line x1="%f" y1="%f" x2="%f" y2="%f" stroke="black" stroke-width="2"/>`, margin, margin, margin, height-margin)

    // Axis labels
    fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="14">x</text>`, width-30, height-margin+20)
    fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="14">y</text>`, margin-20, margin-10)

    // Title
    fmt.Fprintf(f, `<text x="%f" y="%f" font-family="Arial" font-size="18" font-weight="bold">Graphs of Power and Root Functions (Onion Graph)</text>`, margin, margin-30)

    fmt.Fprintln(f, "</svg>")
}
```
